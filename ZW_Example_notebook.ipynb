We'll start by loading in some commonly used Python modules.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

Now let's create and plot some artificial data exhibiting a linear relationship corrupted with noise.

```python
x = np.arange(0,100,1)
y0 = 45+3*x
y = y0+np.random.normal(0,50,len(x))

plt.plot(x,y,'.',label='Noisy data')
plt.plot(x,y0,label='True relationship')
plt.xlabel('x')
plt.ylabel('y')
plt.legend(loc=4)
```
