## Acceleration data
All acceleration datasets were collected at a sample rate of 1707.7Hz. To help with using the data in Jupyter notebooks, the data has been lowpass filtered using a Butterworth filter at 40Hz and then downsampled to a sample rate of 170.77Hz. All data is saved as CSV files.

Let's start by loading the data from the bridge loaded by the mass of a single pedestrian from GitHub.

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
url = "https://raw.githubusercontent.com/zwynne/EDACAB/main/LT2_data/LT2_Bridge_unloaded.csv"

data = pd.read_csv(url,index_col=0,parse_dates=True)
fs = 170.77 # Sampling frequency in Hz
```

The data is organized with each column of the dataframe being the output from an accelerometer channel. The names of the accelerometer channels are shown below.

```python
print(data.columns)
```

Let's plot some of the raw acceleration output. In the output data we can see each of the pedestrian crossings.

```python
fg,ax = plt.subplots(figsize=(12,8))
ax.plot(data['A2Z'])
ax.set_ylabel('Acceleration [$m/s^2$]')
ax.set_xlabel('Date time')
fg.tight_layout()
```

Now let's plot the 30-second Welch averaged power spectral density of each accelerometer channel. We can see that the dynamic behaviour of the bridge is complex, with many peaks in the frequency spectra. We can also see that the data is slightly corrupted by the 50Hz electrical mains signal.

```python
from scipy.signal import welch

fg,ax = plt.subplots(2,sharex='all',figsize=(12,8))

for channel in data.columns:
    f,psd = welch(data[channel],fs=fs,nperseg=30*fs)

    if "Z" in channel:
        ax[0].semilogy(f[1:],psd[1:],label=channel)
    else:
        ax[1].semilogy(f[1:],psd[1:],label=channel)
ax[0].set_title('Vertical response',y=0.9)
ax[1].set_title('Transverse response',y=0.9)
ax[0].set_ylabel('Welch averaged power\nspectral density [$(m/s^2)/Hz^2$]')
ax[1].set_ylabel('Welch averaged power\nspectral density [$(m/s^2)/Hz^2$]')
ax[1].set_xlabel('Frequency [Hz]')
ax[0].legend(loc=1)
ax[1].legend(loc=1)
fg.tight_layout()
```
